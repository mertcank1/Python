```python
kelvin=400
celcius=kelvin-273.15
print(f"{kelvin} Kelvin, {celcius:.2f} Celcius'a eşittir." )
```

```python
r=1.5;
alan= (3.14*(r*r))
cevre=(2*3.14*r)
print(f"{r} cm yarıçaplı dairenin alanı={alan:.2f},çevresi={cevre:.2f}'dir.")
```

```python
yol=500
hiz=110
saat=yol/hiz
tam_saat=int(saat)
dakika=(saat-tam_saat)*60
print(f"{yol}km'lik yolu, saatte {hiz}km hızla {tam_saat}saat {dakika:.0f}'dakikada gidebilirsiniz.")
```
